# Installation of pythtb in anaconda environment

```python
from pythtb import *
import matplotlib. pyplot as plt
# a 1D model 
a_1_lattice_vector = [[1.0]]
a_1_d_orbital = [[0.0]]
my_model = tb_model(1, 1, a_1_lattice_vector, a_1_d_orbital) # spin default = 1 spinless
my_model.set_hop ( -1., 0, 0, [1] )
```

```python
my_model.display()
```

```python
(fig, ax) = my_model.visualize(0, draw_hoppings = 0, ph_color= 'black')
```

```python
(k_vec, k_dist, k_node) = my_model.k_path('fullc', 100)
k_label = [r"$-\pi$", r'$0$', r"$\pi$"]
```

```python
evals = my_model.solve_all(k_vec)
```

```python
fig, ax = plt.subplots()
ax.plot(k_dist,evals[0])
ax.set_title("1D chain band structure")
ax.set_xlabel("Path in k-space")
ax.set_ylabel("Band energy")
ax.set_xticks(k_node)
ax.set_xticklabels(k_label)
ax.set_xlim(k_node[0],k_node[-1])
for n in range(len(k_node)):
  ax.axvline(x=k_node[n], linewidth=0.5, color='k')
fig.tight_layout()
# fig.show()
```

```python
sc_tb = my_model.make_supercell([[2]])
```

```python
(fig, ax) = sc_tb.visualize(0, draw_hoppings = 0, ph_color= 'black')
```

```python

sc_tb = my_model.make_supercell([[8]], to_home=False, to_home_suppress_warning= True)
(k_vec, k_dist, k_node) = sc_tb.k_path([-0.5, 0, 0.5], 1000)
k_label = [r"$-\pi/16$", r'$0$', r"$\pi/16$"]
evals_sc = sc_tb.solve_all(k_vec)
fig, ax = plt.subplots()
ax.plot(k_dist,evals_sc[0])
ax.set_title("1D chain band structure Supercell of 2 unit cells")
ax.set_xlabel("Path in k-space")
ax.set_ylabel("Band energy")
ax.set_xticks(k_node)
ax.set_xticklabels(k_label)
ax.set_xlim(k_node[0],k_node[-1])
for n in range(len(k_node)):
  ax.axvline(x=k_node[n], linewidth=0.5, color='k')
fig.tight_layout()
```

```python
a_1_lattice_vector = [[4.0]]
a_1_d_orbital = [[0.0], [0.25], [0.5], [0.75]]
my_model = tb_model(1, 1, a_1_lattice_vector, a_1_d_orbital) # spin default = 1 spinless
my_model.set_hop ( -1., 0, 1, [0] )
my_model.set_hop ( -1., 1, 2, [0] )
my_model.set_hop ( -1., 2, 3, [0] )
my_model.set_hop ( -1., 3, 0, [1] )
```

```python
(fig, ax) = my_model.visualize(0, draw_hoppings = 1, ph_color= 'black')
```

```python
(k_vec, k_dist, k_node) = my_model.k_path([-2, 0, 2], 1000)
k_label = [r"$-\pi$", r'$0$', r"$\pi$"]
evals = my_model.solve_all(k_vec)
fig, ax = plt.subplots()
for i in range(len(evals)):
  ax.plot(k_dist,evals[i])
ax.set_title("1D chain band structure")
ax.set_xlabel("Path in k-space")
ax.set_ylabel("Band energy")
ax.set_xticks(k_node)
ax.set_xticklabels(k_label)
ax.set_xlim(k_node[0],k_node[-1])
for n in range(len(k_node)):
  ax.axvline(x=k_node[n], linewidth=0.5, color='k')
fig.tight_layout()
# fig.show()
```
